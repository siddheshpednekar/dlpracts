# Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
            
```

```python
from sklearn.datasets import load_boston
data = load_boston()
```

```python
x = data.data
y = data.target
```

```python
x.shape
```

```python
y.shape
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
x_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred= lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
mean_squared_error(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

```python
from sklearn.linear_model import HuberRegressor
hr = HuberRegressor()
```

```python
hr.fit(x_train,y_train)
```

```python
pred= hr.predict(x_test)
pred
```

```python
mean_squared_error(y_test,pred)
```

```python
mean_absolute_error(y_test,pred)
```

```python
r2_score(y_test,pred)*100
```

```python
df = pd.read_csv('diabetes.csv')
```

```python
df.head()
```

```python
x = df.iloc[:,0:8]
x
```

```python
y = df.iloc[:,-1:]
y
```

```python
import keras
```

```python
import tensorflow
```

```python
from keras.layers import Dense
from keras.models import Sequential
```

```python
model = Sequential()
```

```python
model.add(Dense(units=30,input_dim=8,activation='relu'))
model.add(Dense(units=20,activation='relu'))
model.add(Dense(units=10,activation='relu'))
model.add(Dense(units=1,activation='sigmoid'))
```

```python
#binary_crossentropy= binary classification  #Categorical_crossentropy= Multiclass classification 
#optimizer works on Sgd stochastic gradient desecent, to find the minimum value
#metrics = how much decrease in the model
model.compile(loss='binary_crossentropy',optimizer='adam',metrics='accuracy')
```

```python
model.fit(x,y,epochs=1000,batch_size=32)
```
